# **Predicting High-Cost Medicare Beneficiaries: A Machine Learning Comparison Using CMS Synthetic Claims Data**

# **Part 4: Modeling**

In this section, we'll be applying **logistic regression**, **random forest**, and **XGBoost** to answer the following research questions:

Given previous years' information (up to the timestamp of 12/31 of the immediate precedent year), can we predict whether this beneficiary will be high-cost in the upcoming year?

For later stage iterations (during winter break), another research question (model 2) to be tackled would be:

Given previous years' information and 5 months of the current year's information (up to the timestamp of 5/31 of the current claim year), can we predict whether the beneficiary will be projected to high-cost in the current year?

**Parked model 2 right now due to time constraint and model complexity across claims.** Model 2 (out of scope currently) will consider 4 different scenarios: 
| Scenario | History? | First occurrence | Notes                                                              |
| -------- | -------- | ---------------- | ------------------------------------------------------------------ |
| 1a       | No       | Jan–May          | Early in year, no prior info                                       |
| 1b       | No       | Jun–Dec          | Later in year, no prior info                                       |
| 2a       | Yes      | Jan–May          | Early, has history — basically like Model A but partial-year input |
| 2b       | Yes      | Jun–Dec          | Late, has history — partial-year info                              |

```python
import numpy as np
import pandas as pd
import pickle
import os
```

```python
clean_df = pd.read_pickle('model_data/df_model_a.pkl')
clean_df.head(2)
```

# **4.1 Model A Implementation**

```python
from imblearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

# Exclude 2008 (no prior history)
modeling_df = clean_df[clean_df['year'] != 2008].copy()

# Prepare features and target
X = modeling_df.drop(columns=["high_cost", "DESYNPUF_ID", "year"])
y = modeling_df["high_cost"]

# Define features to EXCLUDE (current year costs and attributes because we're not supposed to have these info)
exclude_cols = [
    'high_cost',  
    'DESYNPUF_ID',  
    'year',  
    'MEDREIMB_IP', 'BENRES_IP', 'PPPYMT_IP',
    'MEDREIMB_OP', 'BENRES_OP', 'PPPYMT_OP', 
    'MEDREIMB_CAR', 'PPPYMT_CAR',
    'IP_CLM_UTLZTN_DAY_CNT',
    'OP_claim_duration (days)', 'CAR_claim_duration (days)',
    'PDE_QTY_DSPNSD_NUM'
]

# Also exclude current-year one-hot encoded features (claims THIS year)
current_year_onehot = [col for col in modeling_df.columns 
                       if any(prefix in col for prefix in [
                           'IP_provider', 'IP_ADMTNG', 'IP_PRIMARY', 'IP_UTIL',
                           'OP_provider', 'OP_ADMTNG', 'OP_DGNS1', 'OP_HCPCS',
                           'PDE_NDC_BUCKET', 'CAR_DGNS1', 'CAR_HCPCS'
                       ]) and '_history_sum' not in col]

exclude_cols.extend(current_year_onehot)

X = modeling_df.drop(columns=exclude_cols)
y = modeling_df["high_cost"]

print(f"Features after removing leakage: {X.shape[1]} columns")
print(f"Original data shape: {clean_df.shape}")
print(f"Modeling data shape (excluding 2008): {modeling_df.shape}")
print(f"Class distribution:\n{modeling_df['high_cost'].value_counts(normalize=True)}")
```

```python
# Convert non-numerical values 
X.select_dtypes(include=['object', 'category']).columns
cat_cols = X.select_dtypes(include=['object', 'category']).columns

X = pd.get_dummies(X, columns=cat_cols, drop_first=True)

# Random split since each patient-year is independent
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

X_train = X_train.astype(float)
X_test = X_test.astype(float)
```

```python
# Logistic Regression

pipeline = Pipeline([
    ('smote', SMOTE(random_state=42)),  
    ('lr', LogisticRegression(max_iter=1000, random_state=42))
])

# Grid search
param_grid = {
    'lr__C': [0.01, 0.1, 1, 10],
    'lr__penalty': ['l1', 'l2'],
    'lr__solver': ['liblinear']
}

search = GridSearchCV(
    pipeline,
    param_grid,
    scoring='roc_auc',  
    cv=5,
    n_jobs=-1,
    verbose=1
)

# Fit
print("\nTraining model...")
search.fit(X_train, y_train)

# Evaluate
best_model = search.best_estimator_
y_pred = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[:, 1]

print("\n" + "="*60)
print("RESULTS")
print("="*60)
print("\nBest parameters:", search.best_params_)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nAUC-ROC Score:", roc_auc_score(y_test, y_proba))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt
import numpy as np

# Get probability predictions
y_proba = best_model.predict_proba(X_test)[:, 1]

# Calculate precision-recall at different thresholds
precisions, recalls, thresholds = precision_recall_curve(y_test, y_proba)

# Plot precision-recall tradeoff
plt.figure(figsize=(10, 6))
plt.plot(recalls[:-1], precisions[:-1], linewidth=2)
plt.xlabel('Recall', fontsize=12)
plt.ylabel('Precision', fontsize=12)
plt.title('Precision-Recall Curve', fontsize=14)
plt.grid(True, alpha=0.3)
plt.axhline(y=0.21, color='r', linestyle='--', label='Baseline (current model)')
plt.legend()
plt.show()

# Also plot how metrics change with threshold
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

ax[0].plot(thresholds, precisions[:-1], label='Precision', linewidth=2)
ax[0].plot(thresholds, recalls[:-1], label='Recall', linewidth=2)
ax[0].set_xlabel('Threshold')
ax[0].set_ylabel('Score')
ax[0].set_title('Precision-Recall vs Threshold')
ax[0].legend()
ax[0].grid(True, alpha=0.3)
ax[0].axvline(x=0.5, color='r', linestyle='--', alpha=0.5, label='Default (0.5)')

# Calculate F1 score at each threshold
f1_scores = 2 * (precisions[:-1] * recalls[:-1]) / (precisions[:-1] + recalls[:-1] + 1e-10)
ax[1].plot(thresholds, f1_scores, linewidth=2, color='green')
ax[1].set_xlabel('Threshold')
ax[1].set_ylabel('F1 Score')
ax[1].set_title('F1 Score vs Threshold')
ax[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Find thresholds that give different recall levels
print("\nThreshold analysis for high-cost patient identification:")
print("="*70)
for target_recall in [0.70, 0.75, 0.80, 0.85, 0.90]:
    # Find the threshold that gives closest to target recall
    idx = np.argmin(np.abs(recalls[:-1] - target_recall))
    threshold = thresholds[idx]
    precision = precisions[idx]
    recall = recalls[idx]
    
    # Apply this threshold
    y_pred_temp = (y_proba >= threshold).astype(int)
    
    # Calculate how many patients you'd flag
    n_flagged = y_pred_temp.sum()
    n_true_positive = ((y_pred_temp == 1) & (y_test == 1)).sum()
    n_false_positive = ((y_pred_temp == 1) & (y_test == 0)).sum()
    
    print(f"\nTarget Recall: {target_recall*100:.0f}%")
    print(f"  Threshold: {threshold:.3f}")
    print(f"  Actual Recall: {recall:.3f} (catch {n_true_positive}/{y_test.sum()} high-cost patients)")
    print(f"  Precision: {precision:.3f}")
    print(f"  Total flagged: {n_flagged:,} patients ({n_flagged/len(y_test)*100:.1f}% of population)")
    print(f"  False positives: {n_false_positive:,}")

# Show distribution of predicted probabilities
print("\n" + "="*70)
print("Probability distribution:")
print(f"Min probability: {y_proba.min():.4f}")
print(f"Max probability: {y_proba.max():.4f}")
print(f"Median probability: {np.median(y_proba):.4f}")
print(f"\nPercentiles:")
for p in [10, 25, 50, 75, 90, 95, 99]:
    print(f"  {p}th percentile: {np.percentile(y_proba, p):.4f}")
```

```python
optimal_threshold = 0.514
y_pred_optimal = (y_proba >= optimal_threshold).astype(int)

# Results
print("\n" + "="*60)
print("RESULTS WITH OPTIMIZED THRESHOLD (75% recall target)")
print("="*60)
print(f"\nBest parameters: {search.best_params_}")
print(f"Optimal threshold: {optimal_threshold}")
print(f"\nClassification Report:\n{classification_report(y_test, y_pred_optimal)}")
print(f"\nAUC-ROC Score: {roc_auc_score(y_test, y_proba):.4f}")
print(f"\nConfusion Matrix:\n{confusion_matrix(y_test, y_pred_optimal)}")

# Business metrics
n_flagged = y_pred_optimal.sum()
n_high_cost = y_test.sum()
n_caught = ((y_pred_optimal == 1) & (y_test == 1)).sum()

print(f"\n" + "="*60)
print("BUSINESS METRICS")
print("="*60)
print(f"Total patients: {len(y_test):,}")
print(f"Actual high-cost patients: {n_high_cost:,}")
print(f"Patients flagged for intervention: {n_flagged:,} ({n_flagged/len(y_test)*100:.1f}%)")
print(f"High-cost patients caught: {n_caught:,} ({n_caught/n_high_cost*100:.1f}%)")
print(f"High-cost patients missed: {n_high_cost - n_caught:,} ({(n_high_cost - n_caught)/n_high_cost*100:.1f}%)")
```

```python
# Random forest -- biggest difficulty is in memory
from sklearn.ensemble import RandomForestClassifier

pipeline_rf = Pipeline([
    ('smote', SMOTE(random_state=42)),
    ('rf', RandomForestClassifier(random_state=42, n_jobs=-1))
])

# MUCH smaller grid
param_grid_rf = {
    'rf__n_estimators': [100],           
    'rf__max_depth': [20],               
    'rf__min_samples_split': [5, 10],    
    'rf__min_samples_leaf': [2, 4],      
    'rf__max_features': ['sqrt']         
}

search_rf = GridSearchCV(
    pipeline_rf,
    param_grid_rf,
    scoring='roc_auc',
    cv=3,              
    n_jobs=4,          
    verbose=2
)

print("\nTraining Random Forest (this will take a few minutes)...")
search_rf.fit(X_train, y_train)

# Evaluate
best_rf = search_rf.best_estimator_
y_proba_rf = best_rf.predict_proba(X_test)[:, 1]
y_pred_rf = best_rf.predict(X_test)
```

```python
# After training RF
best_rf = search_rf.best_estimator_
y_proba_rf = best_rf.predict_proba(X_test)[:, 1]
y_pred_rf = best_rf.predict(X_test)

print("\n" + "="*60)
print("RANDOM FOREST RESULTS")
print("="*60)
print("\nBest parameters:", search_rf.best_params_)  
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))  
print("\nAUC-ROC Score:", roc_auc_score(y_test, y_proba_rf))  
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_rf)) 
```

```python
from sklearn.metrics import precision_recall_curve
import numpy as np

# Find threshold for 75% recall
precisions, recalls, thresholds = precision_recall_curve(y_test, y_proba_rf)

print("\nThreshold analysis for Random Forest:")
print("="*70)
for target_recall in [0.70, 0.75, 0.80, 0.85]:
    idx = np.argmin(np.abs(recalls[:-1] - target_recall))
    threshold = thresholds[idx]
    precision = precisions[idx]
    recall = recalls[idx]
    
    # Apply threshold
    y_pred_temp = (y_proba_rf >= threshold).astype(int)
    n_flagged = y_pred_temp.sum()
    n_caught = ((y_pred_temp == 1) & (y_test == 1)).sum()
    
    print(f"\nTarget Recall: {target_recall*100:.0f}%")
    print(f"  Threshold: {threshold:.3f}")
    print(f"  Actual Recall: {recall:.3f}")
    print(f"  Precision: {precision:.3f}")
    print(f"  Patients flagged: {n_flagged:,} ({n_flagged/len(y_test)*100:.1f}%)")
    print(f"  High-cost caught: {n_caught:,}/{y_test.sum()}")

# Apply 75% recall threshold
target_recall = 0.75
idx = np.argmin(np.abs(recalls[:-1] - target_recall))
optimal_threshold_rf = thresholds[idx]
y_pred_rf_optimal = (y_proba_rf >= optimal_threshold_rf).astype(int)

print("\n" + "="*70)
print(f"RANDOM FOREST AT 75% RECALL (threshold={optimal_threshold_rf:.3f})")
print("="*70)
print(classification_report(y_test, y_pred_rf_optimal))
print(f"\nAUC-ROC: {roc_auc_score(y_test, y_proba_rf):.4f}")
```

```python
rf_model = best_rf.named_steps['rf']
feature_importance = pd.DataFrame({
    'feature': X_train.columns,
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=False)

print("\nTop 20 Most Important Features:")
print(feature_importance.head(20))
```

```python
import xgboost as xgb
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
import pandas as pd

# Calculate scale_pos_weight for class imbalance
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()
print(f"Class imbalance ratio: {scale_pos_weight:.2f}")

# Pipeline WITHOUT SMOTE (XGBoost handles imbalance natively)
pipeline_xgb = Pipeline([
    ('xgb', xgb.XGBClassifier(
        random_state=42,
        use_label_encoder=False,
        eval_metric='logloss',
        scale_pos_weight=scale_pos_weight,  
        n_jobs=-1
    ))
])

param_grid_xgb = {
    'xgb__n_estimators': [100, 200],
    'xgb__max_depth': [3, 6],           
    'xgb__learning_rate': [0.1, 0.2],   
    'xgb__subsample': [0.8],            
    'xgb__colsample_bytree': [0.8]      
}

search_xgb = GridSearchCV(
    pipeline_xgb,
    param_grid_xgb,
    scoring='roc_auc',  
    cv=3,
    n_jobs=4,           
    verbose=2
)

# Fit
print("\nTraining XGBoost (this will take a few minutes)...")
search_xgb.fit(X_train, y_train)

# Evaluate
best_xgb = search_xgb.best_estimator_
y_pred_xgb = best_xgb.predict(X_test)
y_proba_xgb = best_xgb.predict_proba(X_test)[:, 1]

print("\n" + "="*60)
print("XGBOOST RESULTS (default threshold)")
print("="*60)
print("\nBest parameters:", search_xgb.best_params_)
print("\nClassification Report:\n", classification_report(y_test, y_pred_xgb))
print("\nAUC-ROC Score:", roc_auc_score(y_test, y_proba_xgb))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_xgb))

# Find 75% recall threshold
from sklearn.metrics import precision_recall_curve
import numpy as np

precisions, recalls, thresholds = precision_recall_curve(y_test, y_proba_xgb)
target_recall = 0.75
idx = np.argmin(np.abs(recalls[:-1] - target_recall))
optimal_threshold_xgb = thresholds[idx]
y_pred_xgb_optimal = (y_proba_xgb >= optimal_threshold_xgb).astype(int)

print("\n" + "="*60)
print(f"XGBOOST AT 75% RECALL (threshold={optimal_threshold_xgb:.3f})")
print("="*60)
print(classification_report(y_test, y_pred_xgb_optimal))

n_flagged = y_pred_xgb_optimal.sum()
n_caught = ((y_pred_xgb_optimal == 1) & (y_test == 1)).sum()
print(f"\nPatients flagged: {n_flagged:,} ({n_flagged/len(y_test)*100:.1f}%)")
print(f"High-cost caught: {n_caught:,}/{y_test.sum()}")

# Feature importance
importance_df_xgb = pd.DataFrame({
    'feature': X_train.columns,
    'importance': best_xgb.named_steps['xgb'].feature_importances_
}).sort_values(by='importance', ascending=False)

print("\n" + "="*60)
print("TOP 20 IMPORTANT FEATURES (XGBoost)")
print("="*60)
print(importance_df_xgb.head(20).to_string(index=False))
```
