```python
def preprocess_beneficiaries(df):

    static_cols = ['SP_STATE_CODE', 'age', 'baseline_age', 'age_bin', 'sex', 'race', 'bene_esrd']
    binary_chronic_cols = [
        'SP_ALZHDMTA','SP_CHF','SP_CHRNKIDN','SP_CNCR','SP_COPD','SP_DEPRESSN',
        'SP_DIABETES','SP_ISCHMCHT','SP_OSTEOPRS','SP_RA_OA','SP_STRKETIA'
    ]
    coverage_cols = ['BENE_HI_CVRAGE_TOT_MONS','BENE_SMI_CVRAGE_TOT_MONS','BENE_HMO_CVRAGE_TOT_MONS','PLAN_CVRG_MOS_NUM']
    cost_cols = ['MEDREIMB_IP','BENRES_IP','PPPYMT_IP','MEDREIMB_OP','BENRES_OP','PPPYMT_OP','MEDREIMB_CAR','BENRES_CAR','PPPYMT_CAR']
    outcome_col = 'is_dead'

    # Convert binary/categorical to numeric
    binary_cols = binary_chronic_cols + [outcome_col, 'bene_esrd']
    df[binary_cols] = df[binary_cols].apply(lambda x: x.astype(int))

    # Build aggregation dictionary
    agg_dict = {c:'first' for c in static_cols}
    agg_dict.update({c:'max' for c in binary_chronic_cols})
    agg_dict.update({c:'sum' for c in coverage_cols})
    agg_dict.update({c:['sum','mean','max'] for c in cost_cols})
    agg_dict[outcome_col] = 'max'

    # Aggregate
    df_agg = df.groupby(['DESYNPUF_ID','year']).agg(agg_dict)

    # Flatten columns
    df_agg.columns = ['_'.join(filter(None,col)).strip() if isinstance(col,tuple) else col for col in df_agg.columns.values]
    df_agg = df_agg.reset_index().sort_values(['DESYNPUF_ID','year'])

    # Lag and delta features
    lag_features = [f"{c}_sum" for c in cost_cols] + [f"{c}_max" for c in binary_chronic_cols]
    for feat in lag_features:
        if feat in df_agg.columns:
            df_agg[f"{feat}_lag"] = df_agg.groupby('DESYNPUF_ID')[feat].shift(1)
            df_agg[f"{feat}_delta"] = df_agg[feat] - df_agg[f"{feat}_lag"]

    return df_agg

```

```python
def aggregate_inpatient(df):
    agg = df.groupby(["DESYNPUF_ID", "year"]).agg({
        "CLM_PMT_AMT": "sum",
        "CLM_ID": "nunique",
        "CLM_UTLZTN_DAY_CNT": ["sum", "mean", "max"],
        "claim_duration (days)": ["mean", "max"],
        "NCH_BENE_IP_DDCTBL_AMT": "sum",
        "NCH_BENE_PTA_COINSRNC_LBLTY_AM": "sum",
        "NCH_BENE_BLOOD_DDCTBL_LBLTY_AM": "sum"
    })

    # Flatten columns
    agg.columns = ["_".join(col).strip() for col in agg.columns.values]

    # Add bucket counts
    bucket_counts = df.groupby(["DESYNPUF_ID", "year", "PRIMARY_ICD9_BUCKET"], observed=True).size().unstack(fill_value=0)
    bucket_counts = bucket_counts.add_prefix("DX_BUCKET_")

    # Admitting diagnosis buckets
    adm_counts = df.groupby(["DESYNPUF_ID", "year", "ADMTNG_BUCKET"], observed=True).size().unstack(fill_value=0)
    adm_counts = adm_counts.add_prefix("ADMT_BUCKET_")

    # Facility type
    facility_counts = df.groupby(["DESYNPUF_ID", "year", "provider_category"], observed=True).size().unstack(fill_value=0)
    facility_counts = facility_counts.add_prefix("FACILITY_")

    # Combine everything
    final = agg.join(bucket_counts, how="left").join(adm_counts, how="left").join(facility_counts, how="left")

    return final.reset_index()
```

```python
def aggregate_outpatient(df):
    """
    Aggregates outpatient claims to the beneficiary-year level.
    
    Parameters
    ----------
    df : pandas.DataFrame
        Outpatient claims with DESYNPUF_ID, year, and precomputed buckets.
        
    Returns
    -------
    pandas.DataFrame
        Wide beneficiary-year table.
    """

    agg_costs = df.groupby(["DESYNPUF_ID", "year"]).agg(
        CLM_PMT_AMT_sum=("CLM_PMT_AMT", "sum"),
        NCH_PRMRY_PYR_CLM_PD_AMT_sum=("NCH_PRMRY_PYR_CLM_PD_AMT", "sum"),
        NCH_BENE_BLOOD_DDCTBL_LBLTY_AM_sum=("NCH_BENE_BLOOD_DDCTBL_LBLTY_AM", "sum"),
        NCH_BENE_PTB_DDCTBL_AMT_sum=("NCH_BENE_PTB_DDCTBL_AMT", "sum"),
        NCH_BENE_PTB_COINSRNC_AMT_sum=("NCH_BENE_PTB_COINSRNC_AMT", "sum"),
        
        num_outpatient_claims=("CLM_ID", "nunique"),
        claim_duration_mean=("claim_duration (days)", "mean"),
        claim_duration_max=("claim_duration (days)", "max"),
    ).reset_index()

    dx_bucket = (
        df.groupby(["DESYNPUF_ID", "year", "DGNS1_BUCKET"], observed=True)
          .size()
          .unstack(fill_value=0)
          .add_prefix("DX_BUCKET_OP_")
          .reset_index()
    )

    admt_bucket = (
        df.groupby(["DESYNPUF_ID", "year", "ADMTNG_BUCKET"], observed=True)
          .size()
          .unstack(fill_value=0)
          .add_prefix("ADMT_BUCKET_OP_")
          .reset_index()
    )

    provider_bucket = (
        df.groupby(["DESYNPUF_ID", "year", "provider_category"], observed=True)
          .size()
          .unstack(fill_value=0)
          .add_prefix("PROVIDER_OP_")
          .reset_index()
    )

    hcpcs_bucket = (
        df.groupby(["DESYNPUF_ID", "year", "HCPCS_BUCKET"], observed=True)
          .size()
          .unstack(fill_value=0)
          .add_prefix("HCPCS_OP_")
          .reset_index()
    )

    merged = (
        agg_costs
        .merge(dx_bucket, on=["DESYNPUF_ID", "year"], how="left")
        .merge(admt_bucket, on=["DESYNPUF_ID", "year"], how="left")
        .merge(provider_bucket, on=["DESYNPUF_ID", "year"], how="left")
        .merge(hcpcs_bucket, on=["DESYNPUF_ID", "year"], how="left")
    )

    # replace any residual NaN from buckets with 0
    bucket_cols = merged.filter(regex="DX_BUCKET_OP_|ADMT_BUCKET_OP_|PROVIDER_OP_|HCPCS_OP_").columns
    merged[bucket_cols] = merged[bucket_cols].fillna(0).astype(int)

    return merged
```

```python
def aggregate_pde(df):
    """
    Aggregates Part D PDE claims to beneficiary-year level.

    Assumes df has:
    DESYNPUF_ID, year, TOT_RX_CST_AMT, PTNT_PAY_AMT,
    QTY_DSPNSD_NUM, DAYS_SUPLY_NUM, NDC_BUCKET
    """
    
    agg = df.groupby(["DESYNPUF_ID", "year"], observed=True).agg(
        TOT_RX_CST_AMT_sum=("TOT_RX_CST_AMT", "sum"),
        PTNT_PAY_AMT_sum=("PTNT_PAY_AMT", "sum"),
        
        num_prescriptions=("PDE_ID", "count"),
        num_unique_ndc_buckets=("NDC_BUCKET", "nunique"),

        QTY_DSPNSD_total=("QTY_DSPNSD_NUM", "sum"),
        QTY_DSPNSD_mean=("QTY_DSPNSD_NUM", "mean"),

        DAYS_SUPLY_total=("DAYS_SUPLY_NUM", "sum"),
        DAYS_SUPLY_mean=("DAYS_SUPLY_NUM", "mean"),
        DAYS_SUPLY_max=("DAYS_SUPLY_NUM", "max"),
    ).reset_index()

    bucket_counts = (
        df.groupby(["DESYNPUF_ID", "year", "NDC_BUCKET"], observed=True)
          .size()
          .unstack(fill_value=0)
          .add_prefix("NDC_BUCKET_")
          .reset_index()
    )

    merged = agg.merge(bucket_counts, on=["DESYNPUF_ID", "year"], how="left")

    bucket_cols = merged.filter(regex="NDC_BUCKET_").columns
    merged[bucket_cols] = merged[bucket_cols].fillna(0).astype(int)

    return merged
```

```python
def aggregate_carrier(df):
    """
    Aggregates carrier (Part B) claims to beneficiary-year level.

    Assumes df contains:
    DESYNPUF_ID, year, CLM_ID, claim_duration (days),
    total_line_nch_payment, total_line_allowed_charge,
    total_line_coinsurance, total_line_deductible,
    total_primary_payer_amount,
    DGNS1_BUCKET, HCPCS_BUCKET
    """

    agg = df.groupby(["DESYNPUF_ID", "year"], observed=True).agg(
        carrier_nch_payment_sum=("total_line_nch_payment", "sum"),
        carrier_allowed_charge_sum=("total_line_allowed_charge", "sum"),
        carrier_coinsurance_sum=("total_line_coinsurance", "sum"),
        carrier_deductible_sum=("total_line_deductible", "sum"),
        carrier_primary_payer_sum=("total_primary_payer_amount", "sum"),
        
        carrier_num_claims=("CLM_ID", "nunique"),
        claim_duration_mean=("claim_duration (days)", "mean"),
        claim_duration_max=("claim_duration (days)", "max"),
    ).reset_index()

    dgns_counts = (
        df.groupby(["DESYNPUF_ID", "year", "DGNS1_BUCKET"], observed=True)
          .size()
          .unstack(fill_value=0)
          .add_prefix("DGNS1_")
          .reset_index()
    )

    hcpcs_counts = (
        df.groupby(["DESYNPUF_ID", "year", "HCPCS_BUCKET"], observed=True)
          .size()
          .unstack(fill_value=0)
          .add_prefix("HCPCS_")
          .reset_index()
    )

    merged = agg.merge(dgns_counts, on=["DESYNPUF_ID", "year"], how="left")
    merged = merged.merge(hcpcs_counts, on=["DESYNPUF_ID", "year"], how="left")

    bucket_cols = merged.filter(regex="^(DGNS1_|HCPCS_)").columns
    merged[bucket_cols] = merged[bucket_cols].fillna(0).astype(int)

    return merged
```

```python
def validate_medicare_dataset_by_year(df, year_col="year"):
    """
    Create validation summary tables grouped by year:
      A1. Demographics by Year
      B1. Chronic Conditions by Year
      C1. Cost Summary by Year
      D1. Utilization Summary by Year
      E2. State-Level Spending (Ranked) by Year
    """
    results = {}

    age_col = "age_first"
    sex_col = "sex_first"
    race_col = "race_first"
    state_col = "SP_STATE_CODE_first"

    chronic_cols = [
        'SP_ALZHDMTA_max','SP_CHF_max','SP_CHRNKIDN_max','SP_CNCR_max','SP_COPD_max',
        'SP_DEPRESSN_max','SP_DIABETES_max','SP_ISCHMCHT_max','SP_OSTEOPRS_max',
        'SP_RA_OA_max','SP_STRKETIA_max'
    ]

    cost_cols = [
        'MEDREIMB_IP_sum','BENRES_IP_sum','PPPYMT_IP_sum',
        'MEDREIMB_OP_sum','BENRES_OP_sum','PPPYMT_OP_sum',
        'MEDREIMB_CAR_sum','BENRES_CAR_sum','PPPYMT_CAR_sum', 
        'TOT_RX_CST_AMT_sum', 'PTNT_PAY_AMT_sum'
    ]

    # Ensure year is present
    if year_col not in df.columns:
        raise ValueError(f"Dataset is missing the year column: {year_col}")

    # A. Demographics by Year
    demo_list = []

    for yr, grp in df.groupby(year_col):
        race_dist = grp[race_col].value_counts(normalize=True) * 100  # percentages
        race_dist = race_dist.reindex(['Black', 'Hispanic', 'Other', 'White'], fill_value=0)
    
        demo_list.append({
            "year": yr,
            "Black": race_dist['Black'],
            "Hispanic": race_dist['Hispanic'],
            "Other": race_dist['Other'],
            "White": race_dist['White'],
            "mean_age": grp[age_col].mean(),
            "median_age": grp[age_col].median(),
            "%_male": (grp[sex_col] == "Male").mean() * 100,
            "%_female": (grp[sex_col] == "Female").mean() * 100,
            "n": len(grp)
        })

    results["Demographics_by_Year"] = pd.DataFrame(demo_list).sort_values("year")


    for yr, grp in df.groupby(year_col):
        demo_list.append({
            "year": yr,
            "race": race_col,
            "mean_age": grp[age_col].mean(),
            "median_age": grp[age_col].median(),
            "%_male": (grp[sex_col] == "Male").mean() * 100,
            "%_female": (grp[sex_col] == "Female").mean() * 100,
            "n": len(grp)
        })

    results["Demographics_by_Year"] = pd.DataFrame(demo_list).sort_values("year")

    # B. Chronic Condition Prevalence by Year
    chronic_df = df.groupby(year_col)[chronic_cols].mean().reset_index()
    chronic_df[chronic_cols] = chronic_df[chronic_cols] * 100  # convert only chronic cols to %
    chronic_df = chronic_df.sort_values(year_col)
    results["Chronic_Prevalence_by_Year"] = chronic_df


    # C. Cost Summary by Year
    cost_year = df.groupby([year_col, state_col])["total_payer_cost"].agg(["mean", "median", "sum", "std"])

    results["Cost_Summary_by_Year"] = cost_year

    # D. Utilization Summary by Year
    util_list = []

    # Define inpatient indicator
    df["had_inpatient"] = df["MEDREIMB_IP_sum"] > 0

    for yr, grp in df.groupby(year_col):
        util_list.append({
            "year": yr,
            "inpatient_rate_%": grp["had_inpatient"].mean() * 100,
            "n": len(grp)
        })

    results["Utilization_by_Year"] = pd.DataFrame(util_list).sort_values("year")

    # E. State-Level Spending Ranked by Year
    df["total_payer_cost"] = df[cost_cols].sum(axis=1)

    state_year = (
        df.groupby([year_col, state_col])["total_payer_cost"]
        .agg(["sum", "mean", "count"])
        .sort_values(["year", "sum"], ascending=[True, False])
        .reset_index()
    )

    results["State_Spending_by_Year"] = state_year

    return results
```

```python
def validate_medicare_dataset_by_year_pretty(df, year_col="year"):
    results = validate_medicare_dataset_by_year(df, year_col)

    styled_results = {}

    # ---- Style 1: Demographics ----
    styled_results["Demographics_by_Year"] = (
        results["Demographics_by_Year"]
        .style.format({
            "Black": "{:.1f}%",
            "Hispanic": "{:.1f}%",
            "Other": "{:.1f}%",
            "White": "{:.1f}%",
            "mean_age": "{:.1f}",
            "median_age": "{:.1f}",
            "%_male": "{:.1f}%",
            "%_female": "{:.1f}%",
            "n": "{:,}"
        })
        .set_caption("Demographics by Year")
        .hide(axis="index")
        .background_gradient(cmap="Blues")
    )

    # ---- Style 2: Chronic Condition Prevalence (row-wise) ----
    styled_results["Chronic_Prevalence_by_Year"] = (
        results["Chronic_Prevalence_by_Year"]
        .style.format("{:.1f}%", subset=chronic_cols)  # format only chronic columns
        .set_caption("Chronic Condition Prevalence by Year (%)")
        .hide(axis="index")
        .background_gradient(cmap="Reds", subset=chronic_cols, axis=None)  # color only chronic columns


    )

    # ---- Style 3: Cost Summary ----
    styled_results["Cost_Summary_by_Year"] = (
        results["Cost_Summary_by_Year"]
        .style.format("{:,.0f}")
        .set_caption("Cost Summary by Year")
        .background_gradient(cmap="Greens")
    )

    # ---- Style 4: Utilization ----
    styled_results["Utilization_by_Year"] = (
        results["Utilization_by_Year"]
        .style.format({
            "inpatient_rate_%": "{:.1f}%",
            "n": "{:,}"
        })
        .set_caption("Utilization Summary by Year")
        .hide(axis="index")
        .background_gradient(cmap="Purples")
    )

    # ---- Style 5: State spending (row-wise) ----
    styled_results["State_Spending_by_Year"] = (
        results["State_Spending_by_Year"]
        .style.format({
            "sum": "{:,.0f}",
            "mean": "{:,.0f}",
            "count": "{:,}"
        })
        .set_caption("State-Level Spending by Year (Ranked)")
        .background_gradient(
            cmap="Oranges",
            subset=results["State_Spending_by_Year"].columns.difference(["year"]),
            axis=1  # gradient computed across each row (per year)
    )
)

    return styled_results
```

```python
def print_validation_tables(df):
    """Run validation and print all results as styled tables."""

    # Run the core validation function
    results = validate_medicare_dataset_by_year(df)

    # ---- 1. Demographics ----
    print("\n DEMOGRAPHICS BY YEAR ===\n")
    results["Demographics_by_Year"] = results["Demographics_by_Year"].drop(columns=["race"], errors="ignore")
    results["Demographics_by_Year"] = results["Demographics_by_Year"].dropna(subset=["Black"]).reset_index(drop=True)
    for col in ["Black", "Hispanic", "Other", "White", "mean_age", "median_age", "%_male", "%_female"]:
        results["Demographics_by_Year"][col] = results["Demographics_by_Year"][col].astype(float)

    demo = results["Demographics_by_Year"].style.format({
        "Black": "{:.1f}%",
        "Hispanic": "{:.1f}%",
        "Other": "{:.1f}%",
        "White": "{:.1f}%",
        "mean_age": "{:.1f}",
        "median_age": "{:.1f}",
        "%_male": "{:.1f}%",
        "%_female": "{:.1f}%",
        "n": "{:,}"
    }).background_gradient(cmap="Blues",
                           subset=["Black", "Hispanic", "Other", "White", "mean_age", "median_age", "%_male", "%_female", "n"]
)
    display(demo)

    # ---- 2. Chronic condition prevalence ----
    print("\n CHRONIC CONDITION PREVALENCE BY YEAR (%) \n")
    chronic_cols = [
        'SP_ALZHDMTA_max','SP_CHF_max','SP_CHRNKIDN_max','SP_CNCR_max','SP_COPD_max',
        'SP_DEPRESSN_max','SP_DIABETES_max','SP_ISCHMCHT_max','SP_OSTEOPRS_max',
        'SP_RA_OA_max','SP_STRKETIA_max'
    ]
    chronic_cols_subset = chronic_cols  # only the chronic columns

    chronic = results["Chronic_Prevalence_by_Year"].style.format(
            "{:.1f}%", subset=chronic_cols_subset
        ).background_gradient(
            cmap="Reds", subset=chronic_cols_subset, axis=None
        )
    display(chronic)

    # ---- 3. Cost summary ----
    print("\n COST SUMMARY BY YEAR \n")
    cost = results["Cost_Summary_by_Year"].style.format("{:,.0f}") \
        .background_gradient(cmap="Greens")
    display(cost)

    # ---- 4. Utilization summary ----
    print("\n UTILIZATION SUMMARY BY YEAR \n")
    util = results["Utilization_by_Year"].style.format({
        "inpatient_rate_%": "{:.1f}%",
        "op_visits_mean": "{:.1f}",
        "op_visits_median": "{:.1f}",
        "n": "{:,}"
    }).background_gradient(cmap="Purples")
    display(util)

    # ---- 5. State-level spending ----
    print("\n STATE-LEVEL SPENDING BY YEAR (RANKED) \n")
    state = results["State_Spending_by_Year"].style.format({
        "sum": "{:,.0f}",
        "mean": "{:,.0f}",
        "count": "{:,}"
    }).background_gradient(cmap="Oranges")
    display(state)


# ---- Call this to print everything nicely ----
print_validation_tables(final_df)
```

# Data Validation Reflection: De-SynPUF Medicare Dataset

## State-Level Spending
- Across 2008–2010, the synthetic dataset shows the highest HHA claim counts consistently for:
  - California → Florida → Texas (e.g., 2008 counts: CA = 5, FL = 10, TX = 45)
- Comparing with actual Medicare HHA spending:
  - 2008: TX $2.67B > FL $2.35B > CA $1.3B
  - 2010: TX $3.07B > FL $2.28B > CA $1.38B
- Notes:
  1. De-SynPUF includes **inpatient, outpatient, carrier, and PDE claims**, not just HHA.
  2. The data is a **1/20 sample of the 5% sample of CMS claims**, then syntheticized, so exact dollar values differ.

## Demographics
- Gender distribution in the synthetic dataset matches reality: **43% male, 57% female**.
- Race distribution is also reasonably aligned:
  - White ~84% vs ~85% actual
  - Black ~10% vs ~11% actual

## Utilization
- 2008 patient counts and visits highlight the same high-utilization states:
  - Texas: ~332k patients, 20.6M visits
  - Florida: ~321k patients, 22.8M visits
  - California: ~236k patients, 7.8M visits
- The synthetic dataset reflects these patterns with claim counts: CA = 9,631, FL = 7,407, TX = 6,357. Utilization patterns are approximated using the sum of claims payments per category, which correlates with the number of services received.

## Chronic Conditions
- De-SynPUF overestimates some chronic condition prevalence compared to real data:
  - Alzheimer’s/dementia: ~21% vs ~8% actual
  - Diabetes: ~42% vs ~23% actual
- This is **intentional**, to maintain realistic correlations between conditions, costs, utilization, and drug prescriptions.

## Reflection / Conclusion
- Although absolute values differ from real-world Medicare data, **the distributions and trends are consistent**, making De-SynPUF relevant for modeling and analytics.  
- The dataset preserves **relationships between variables**, supports research use, and maintains **patient privacy** through its synthetic design.  
- Overall, despite some differences in prevalence and exact values, the dataset provides a **trustworthy environment for modeling Medicare spending, utilization, and demographics**.

```python
top10_cutoffs = final_df.groupby('year')['total_payer_cost'].quantile(0.90)
print("Top 10% cost cutoff per year:")
print(top10_cutoffs)

for yr in [2008, 2009, 2010]:
    print(f"\n=== Year {yr} ===")
    
    # Filter data for this year
    year_df = final_df[final_df['year'] == yr].copy()
    
    # Flag high-cost patients
    cutoff = top10_cutoffs[yr]
    year_df['high_cost'] = year_df['total_payer_cost'] > cutoff
    
    # --- 1. Demographics ---
    demo_cols = ["age_first", "sex_first", "race_first"]
    demo_summary = year_df.groupby('high_cost')[demo_cols].describe()
    print("\nDemographics Summary:")
    display(demo_summary)
    
    # --- 2. Chronic Conditions ---
    chronic_cols = [
        'SP_ALZHDMTA_max','SP_CHF_max','SP_CHRNKIDN_max','SP_CNCR_max','SP_COPD_max',
        'SP_DEPRESSN_max','SP_DIABETES_max','SP_ISCHMCHT_max','SP_OSTEOPRS_max',
        'SP_RA_OA_max','SP_STRKETIA_max'
    ]
    chronic_summary = year_df.groupby('high_cost')[chronic_cols].mean() * 100  # % prevalence
    print("\nChronic Condition Prevalence (%):")
    display(chronic_summary)
    
    # --- 3. Utilization Patterns ---
    # Assuming you have visit counts columns if needed; otherwise, can sum cost as proxy
    util_cols = [
        'MEDREIMB_IP_sum','BENRES_IP_sum','PPPYMT_IP_sum',
        'MEDREIMB_OP_sum','BENRES_OP_sum','PPPYMT_OP_sum',
        'MEDREIMB_CAR_sum','BENRES_CAR_sum','PPPYMT_CAR_sum', 'TOT_RX_CST_AMT_sum', 'PTNT_PAY_AMT_sum'
    ]
    util_summary = year_df.groupby('high_cost')[util_cols].mean()
    print("\nUtilization Patterns (mean $ by type):")
    display(util_summary)
    
    # --- 4. Cost Breakdown ---
    cost_cols = [
        'MEDREIMB_IP_sum','BENRES_IP_sum','PPPYMT_IP_sum',
        'MEDREIMB_OP_sum','BENRES_OP_sum','PPPYMT_OP_sum',
        'MEDREIMB_CAR_sum','BENRES_CAR_sum','PPPYMT_CAR_sum', 
        'TOT_RX_CST_AMT_sum', 'PTNT_PAY_AMT_sum'
    ]
    cost_summary = year_df.groupby('high_cost')[cost_cols].sum()
    print("\nCost Breakdown (total $):")
    display(cost_summary)
    
    # --- Optional: Visualization ---
    plt.figure(figsize=(8,5))
    sns.histplot(year_df, x='total_payer_cost', hue='high_cost', bins=50, kde=True, palette=['skyblue','orange'])
    plt.axvline(cutoff, color='red', linestyle='--', label=f'Top 10% cutoff = ${cutoff:,.0f}')
    plt.title(f'Annual Cost Distribution - {yr}')
    plt.xlabel('Annual Cost ($)')
    plt.ylabel('Number of Beneficiaries')
    plt.legend()
    plt.show()
```

1. Using **15%** ($11,000+ annual cost) captured too large a portion of patients on the zoomed-in cost distribution curve.
   
3. Using **5%** ($23,000+ annual cost) felt too aggressive, excluding many patients who were still meaningfully high-cost.

## Overall Insight
Across all three years, **high-cost patients consistently have more chronic conditions and higher service utilization** than non-high-cost patients. While absolute costs and some prevalence percentages vary (especially in 2010), the **relative patterns remain robust**, validating the use of the **DE-SynPUF dataset for modeling high-cost patient behavior.** All 3 years are 10% high-cost and 90% non-high-cost, posing moderate imbalance which means we should prioritize precision, recall, and F1-score instead of accuracy for evaluating model outputs. It will also be helpful to maybe conduct SMOTE to help put more emphasis in the high-cost patient predicitions.


```python
top10_cutoffs = {
    2008: 16574.0,
    2009: 17279.4,
    2010: 9830.0
}

# Create a new column 'high_cost' based on year-specific cutoff
final_df['high_cost'] = final_df.apply(
    lambda row: 1 if row['total_payer_cost'] >= top10_cutoffs[row['year']] else 0,
    axis=1
)
```
